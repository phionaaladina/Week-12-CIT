Question 4
Create a dataframe with 10 rows and 3 columns, where the values are random numbers between 1 and 10 (inclusive). Then, replace all values greater than 5 with the value 10.

```python
random_df = pd.DataFrame(np.random.randint(0,11,size=(10, 3)), columns=list('XYZ'))
random_df
```

```python
#REPLACE
random_df.loc[random_df['X'] > 5] = 10
random_df.loc[random_df['Y'] > 5] = 10
random_df.loc[random_df['Z'] > 5] = 10
random_df
```
